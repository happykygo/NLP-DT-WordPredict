```python
# file = open("wordTest.csv", "r")
# lines = file.readlines()
# lines

# get file content into DF
df = pd.DataFrame(open("wordTest.csv", "r").readlines(), columns = ['ORG_DATA'])
df

# split 3 columns
df[['Target', 'Position', 'Sentence']] = df.ORG_DATA.apply(lambda x: pd.Series(str(x).split(" ", 2)))
df
```

```python
def _splitFeatures(sent, posi, rang):
    sent = str(sent).split(" ")
    str_len = len(sent)
    list = [sent[int(posi)-i] if int(posi)-i >=0 else "na" for i in range(rang, 0, -1)] \
+ [sent[int(posi)+1+i] if int(posi)+1+i < str_len else "na" for i in range(rang)]
    return "|".join(list)
```

```python
df[['Words_List']] = pd.DataFrame(df.apply(lambda x : _splitFeatures(x['Sentence'], x['Position'], 2), axis = 1))
df
```

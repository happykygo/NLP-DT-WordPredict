```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from decisiontree import DecisionTreeModel as DT, accuracy_score
import utils as ut
```

## Learning Curve

```python
def one_fold(frac):    
    n = 3 if frac < 1.0 else 1
    accuracy = 0   
    for i in range(n):
        X, y, vocabs_oh = ut.create_dataset("hw1_dataset/hw1.train_mid_rnd.col", 1, frac=frac)

        model = DT()
        model.fit(X,y)
        
        X_test, y_test, _ = ut.create_dataset("hw1_dataset/hw1.test.col", 1, 'both', vocabs_oh)
        y_test_pred = model.predict(X_test)
        
        acc = accuracy_score(y_test, y_test_pred)
        accuracy += acc
        
    return accuracy / n
```

```python
acc = one_fold(0.1)
```

```python
acc
```

```python
fractions = [0.1, 0.2, 0.5, 0.8, 1.0]
```

```python
accuracies = [one_fold(frac) for frac in fractions]
```

```python
accuracies
```

```python
plt.plot(fractions, accuracies)
plt.title('Learning Curve')
plt.xlabel('Fraction of Samples')
plt.ylabel('Accuracy')
```

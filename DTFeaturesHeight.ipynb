```python
import pandas as pd
import numpy as np
import random
from decisiontree import DecisionTreeModel as DT, accuracy_score
import utils as ut
```

## Model1 -- 1 word before target

```python
X, y, vocabs_oh = ut.create_dataset("hw1_dataset/hw1.train_mid_rnd.col", 1, 'before')
X
```
